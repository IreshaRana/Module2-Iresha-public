## 3. Data read/write with i/loc
'iloc' method in pandas is used for index-based referencing whereas 'loc' is used for label-based indexing. 
### Data read with iloc

```python
# Create a dataframe
data = {
    'Name':['Alice','Bob','Charlie','David'],
    'Age':[25,30,35,40],
    'City':['New York','Los Angeles','Chicago','Houston']
    }

df=pd.DataFrame(data)
```

#### Select a row

```python
# Select a row
row=df.iloc[0]
row
```

#### Select a column

```python
# Select a column
column=df.iloc[:,1]
column
```

### Data read with loc
#### Select a row

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

# Create a dataframe with row labels
df=pd.DataFrame(data,index=['a','b','c','d'])
df
```

```python
# Select row label 
row=df.loc['b']
row
```

### Merge dataframes
Merge based on common columns or indices

```python
# Data
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['B', 'C', 'D'], 'Value2': [4, 5, 6]})
df2
```

```python
# Inner join on 'Key'
result = pd.merge(df1, df2, on='Key', how='outer')

print(result)
```

### Join
To join based on indices.

```python
# Data
df1 = pd.DataFrame({'A': [1, 2]}, index=['a', 'b'])
df2 = pd.DataFrame({'B': [3, 4]}, index=['a', 'c'])
df2
```

```python
# Left join on index
result = df1.join(df2, how='left')

print(result)
```

## 6. Other data manipulations

### Data subsetting
Allows to filter and select specific portions of your DataFrame based on conditions or criteria
#### Filter row by applying a condition to columns

```python
# Data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}
df = pd.DataFrame(data)

# Subset rows where age is greater than 30
subset_df=df[df['Age']>30]

subset_df
```

### Data sampling

```python
# Using sample() method

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 35, 40, 45]}
df = pd.DataFrame(data)

# Randomly sample 3 rows
sample_df = df.sample(n=3, random_state=1)

print(sample_df)
```

```python
# Using groupby() - for strateified sampling, where you want to sample proportionally from each group

# Data 
data = {'Group': ['A', 'A', 'A', 'B', 'B', 'B'],
        'Value': [10, 20, 30, 40, 50, 60]}
df = pd.DataFrame(data)

# Stratified sampling: sample 50% from each group
stratified_sample = df.groupby('Group', group_keys=False).sample(frac=0.5, random_state=1)

print(stratified_sample)
```

### Sorting

```python
# Sort by columns

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}
df = pd.DataFrame(data)

# Sort by 'Age' in ascending order
sorted_df = df.sort_values(by='Age')

print(sorted_df)
```

```python
# Sort by index

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}
df = pd.DataFrame(data, index=[3, 1, 2, 0])
df

# Sort rows by index in ascending order
sorted_df = df.sort_index()

print(sorted_df)
```

### Add / delete indexes

```python
# Add index

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# Add a new index column
df['Index'] = ['a', 'b', 'c']

# Set 'NewIndex' column as the index
df_with_new_index = df.set_index('Index')

print(df_with_new_index)
```

```python
# Reset index

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])

# Reset the index and drop the old index
df_reset = df.reset_index(drop=True)

print(df_reset)
```

```python
# Drop index, by converting to a form that does not use an index like an numpy array

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])

# Convert DataFrame to a NumPy array
array = df.values

print(array)
```

### Muti indexes: add, remove

```python
# Define a MultiIndex using tuples
index = pd.MultiIndex.from_tuples(
    [('A', 1), ('A', 2), ('B', 1), ('B', 2)],
    names=['Letter', 'Number']
)

# Create a DataFrame with the MultiIndex
df = pd.DataFrame({
    'Value': [10, 20, 30, 40]
}, index=index)

print(df)
```

### pivot tables

```python
# DataFrame
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 20, 30, 40]
}
df = pd.DataFrame(data)

# Create a pivot table
pivot_table = pd.pivot_table(df, values='Value', index='Date', columns='Category', aggfunc='sum')

print(pivot_table)
```

### idxmin, idxmax
idxmin - to identify the position of the smallest value in your data

```python
# Data
s = pd.Series([10, 20, 5, 15])

# Find the index of the minimum value
index_of_min = s.idxmin()

print(index_of_min)
```

```python
# Find the index of the maximum value
index_of_max = s.idxmax()

print(index_of_max)
```

### value_counts

value_counts - to count the occurrences of unique values in a Series

```python
# Data
s = pd.Series(['apple', 'banana', 'apple', 'orange', 'banana', 'banana'])

# Get the count of each unique value
value_counts = s.value_counts()

print(value_counts)
```

### unique
To find the unique values in a Series or DataFrame column

```python
# Data
s = pd.Series(['apple', 'banana', 'apple', 'orange', 'banana'])

# Get unique values
unique_values = s.unique()

print(unique_values)
```
