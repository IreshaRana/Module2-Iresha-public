## Line charts

```python
# First let's create an np array from 0 to 10 with a 0.1 interval
import numpy as np
x = np.linspace(0,10,101)
x
```

```python
# Then let's calculate sin and cos of x
y = np.cos(x)
z = np.sin(x)
```

### Basic line chart

```python
# Now let's plot a line chart
plt.plot(x,y)
```

### Line chart with additional features

```python
# Add linewidth
plt.plot(x,y,linewidth=10)
```

```python
# Change line pattern
plt.plot(x,y,ls='--')  # If ls is set to 'solid' it will give a continuous line
```

### Draw more than one line in a chart

```python
# Draw two plots at once with different line patterns
plt.plot(x,y,'--',x**2,y**2,'-.')
```

```python
## Drawing two lines after changing line properties

# Draw two lines
lines = plt.plot(x,y,x,z)

# Set properties for the lines using plt.setp()
plt.setp(lines, color='g', linewidth=5)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('sin and cos of value')
plt.title('Plot of sin and cos')

# Show the plot
plt.show()
```

## Scatter plots

```python
# Define data
x = [3,5,2,6,4,6]
y = [6,7,8,6,8,7]

# Create a scatter plot
plt.scatter(x,y)

# Add labels and title
plt.xlabel('number')
plt.ylabel('age')
plt.title('Age distribution according to number')
```

## Process (in Datacamp cheat sheet) in action

```python
# Step 1 - Prepare data
x = [1,2,3,4]
y = [10,20,25,30]

# Step 2 - Create plot, i.e. for this example, create a figure
fig = plt.figure()

# And add a single subplot to the fiture
ax = fig.add_subplot(111)  # '111' means 1 row, 1 column, first (and only) subplot

# Step 3 - Plot
ax.plot(x,y)


# Step 4 - Cutomise plot, i.e. for this example, adding labels and title
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('Single subplot example')

# Step 5 - Save plot
plt.savefig('Fig1')  # The figure will be saved in the current working directory. 

# Step 6 - Show plot
plt.show()
```

### Stacked bar chart

```python
# Data
categories = ['Q1', 'Q2', 'Q3', 'Q4']
apples = [10, 15, 20, 25]
bananas = [5, 10, 15, 20]

# Create a stacked bar chart
plt.bar(categories, apples, label='apples', color='r')
plt.bar(categories, bananas, label='banana', color='y')
```
